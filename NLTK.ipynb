```python
# Sentence tokenizing
# 문단 => 문장
```

```python
from nltk.tokenize import sent_tokenize
# gutenburg corpus
raw_text = gb.raw(fileids='austen-emma.txt')
```

```python
# sentence tokenizer
len(sent_tokenize(raw_text))
```

```python
sents = sent_tokenize(raw_text)
```

```python
# word tokenizing
# sentence => words
from nltk.tokenize import word_tokenize
```

```python
word_tokenize(sents[0])
```

```python
# word punctTokenizer
from nltk.tokenize import WordPunctTokenizer
wpt = WordPunctTokenizer()
wpt.tokenize(sents[0])
```

NLTK Stemming

```python
from nltk.stem import PorterStemmer
import numpy as np
```

```python
tokens = word_tokenize(sents[0])
```

```python
ps = PorterStemmer()
ps.stem(tokens[1])

#words stemming #1 
stem_tokens = []
for token in tokens:
    stem_token = ps.stem(token)
    stem_tokens.append(stem_token)

# python comphrehension #2
stem_tokens = [ps.stem(token) for token in tokens]
stem_tokens
```

```python
# 바뀐 것들 확인 
tokens[1], stem_tokens[1]
```

```python
# stemming을 통해 단어가 바뀐것만 추출해서 확인
mask = [token != stem_token for token, stem_token in zip(tokens, stem_tokens)]
```

```python
masked_tokens = np.array(tokens)[mask]
masked_stem_tokens = np.array(stem_tokens)[mask]
```

```python
np.stack([masked_tokens, masked_stem_tokens], -1)
```
